# Notebook 02 — NLP Preprocessing Walkthrough

This notebook walks through every preprocessing step one-by-one so you can see exactly what changes.

```python
import sys; sys.path.insert(0,'..')
import pandas as pd
from src.data.load_data import load_raw_data
from src.data.preprocess import *

df = load_raw_data()
spam_sample = df[df['label']=='spam']['message'].iloc[0]
ham_sample  = df[df['label']=='ham']['message'].iloc[5]
print('SPAM:', spam_sample)
print('HAM: ', ham_sample)
```

```python
# Step-by-step on spam sample
text = spam_sample
print('1. Original        :', text[:80])
text = to_lowercase(text)
print('2. Lowercase       :', text[:80])
text = remove_html_tags(text)
print('3. No HTML         :', text[:80])
text = remove_urls(text)
print('4. URLs removed    :', text[:80])
text = remove_emails(text)
print('5. Emails removed  :', text[:80])
text = remove_phone_numbers(text)
print('6. Phones removed  :', text[:80])
text = remove_special_characters(text)
print('7. Clean chars     :', text[:80])
tokens = tokenize(text)
print('8. Tokens          :', tokens[:15])
tokens = remove_stopwords(tokens)
print('9. No stopwords    :', tokens[:15])
tokens = stem_tokens(tokens)
print('10. Stemmed        :', tokens[:15])
print('11. FINAL          :', tokens_to_string(tokens))
```

```python
# Compare spam vs ham after full pipeline
print('SPAM clean:', preprocess_text(spam_sample))
print()
print('HAM  clean:', preprocess_text(ham_sample))
```
